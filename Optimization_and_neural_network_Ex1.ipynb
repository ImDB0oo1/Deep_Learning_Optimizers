#Import libraries

```python
import numpy as np

#make dataset
from sklearn.datasets import make_classification

#mlp model
from sklearn.neural_network import MLPClassifier

#evaluation metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

#ploting the results
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

```

#Generate 5 defferent datasets

```python
# Generate a random n-class classification problem with 100 samples
X_1, y_1 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=3, random_state=1)
X_2, y_2 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=3, random_state=2)
X_3, y_3 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=4, random_state=10)
X_4, y_4 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=4, random_state=14)
X_5, y_5 = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=4, random_state=5)
```

#Create classification report function

```python
# create classification report for model evaluation
def create_report(y_test, y_pred):
    print(classification_report(y_test, y_pred))
    # plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

#Apply MLP models on datasets

##Model 1

###Fitting the data

```python
mlp_clf_1 = MLPClassifier(hidden_layer_sizes=(150,100,50),
                        max_iter = 300,activation = 'relu',
                        solver = 'adam')

mlp_clf_1.fit(X_1, y_1)
```

###Evalution of model

```python
# create report 
create_report(y_1, mlp_clf_1.predict(X_1))
```

###Checking the boundries

```python
# Create a grid of points that covers the input feature space
x_min, x_max = X_1[:, 0].min() - 0.1, X_1[:, 0].max() + 0.1
y_min, y_max = X_1[:, 1].min() - 0.1, X_1[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))
# Use the trained MLP to make predictions on the grid of points
Z = mlp_clf_1.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
# Plot the decision boundary that separates the predicted classes
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X_1[:, 0], X_1[:, 1], c=y_1, cmap=plt.cm.RdYlBu, edgecolor='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
